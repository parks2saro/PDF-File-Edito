```python
df_bk = df.drop(index = [0,1]).copy()
df_bk = df_bk.reset_index(drop=True)
df_bk
```

```python
# ADD: FRONT OUTER & FRONT INNER
if df_bk[bookname][0] == 1:
    y = "COVER_FRONT OUTER"
    z = "COVER_FRONT INNER"
    line1 = pd.DataFrame({"CONTENTS":[z],bookname: [0]}, index=[2.5])
    line2 = pd.DataFrame({"CONTENTS":[y],bookname: [0]}, index=[2.6])
    df_bk = pd.concat([df_bk.iloc[:2],line1,df_bk.iloc[2:]]).reset_index(drop=True)
    df_bk = pd.concat([df_bk.iloc[:2],line2,df_bk.iloc[2:]]).reset_index(drop=True)                

    
if df_bk[bookname][0] > 1:
    x = df_bk[bookname][0]
    print(x)
    for i in reversed(range(x)):
        print(i)
        y = "COVER_FRONT OUTER "+ str(i+1)
        z = "COVER_FRONT INNER "+ str(i+1)
        line1 = pd.DataFrame({"CONTENTS":[z],bookname: [0]}, index=[i+2.5])
        line2 = pd.DataFrame({"CONTENTS":[y],bookname: [0]}, index=[i+2.6])
        df_bk = pd.concat([df_bk.iloc[:2],line1,df_bk.iloc[2:]]).reset_index(drop=True)
        df_bk = pd.concat([df_bk.iloc[:2],line2,df_bk.iloc[2:]]).reset_index(drop=True)                

df_bk
```

```python
# move END 
y = df_bk.index[df_bk['CONTENTS'].str.contains('END')].tolist()
print (y)
idx = df_bk.index.tolist()
idx = idx + y
for i in reversed(y):
    idx.pop(i)
    print(idx)
df_bk = df_bk.reindex(idx)
df_bk = df_bk.reset_index(drop=True)
df_bk
```

```python
# ADD: BACK INNER & BACK OUTER

if df_bk[bookname][1] == 1:
    y = "COVER_BACK INNER"
    z = "COVER_BACK OUTER"
    line1 = pd.DataFrame({"CONTENTS":[y],bookname: [0]})
    line2 = pd.DataFrame({"CONTENTS":[z],bookname: [0]})
    df_bk = pd.concat([df_bk,line1]).reset_index(drop=True)
    df_bk = pd.concat([df_bk,line2]).reset_index(drop=True)                

    
if df_bk[bookname][1] > 1:
    x = df_bk[bookname][0]
    print(x)
    for i in range(x):
        print(i)
        y = "COVER_BACK INNER "+ str(i+1)
        z = "COVER_BACK OUTER "+ str(i+1)
        line1 = pd.DataFrame({"CONTENTS":[y],bookname: [0]})
        line2 = pd.DataFrame({"CONTENTS":[z],bookname: [0]})
        df_bk = pd.concat([df_bk,line1]).reset_index(drop=True)
        df_bk = pd.concat([df_bk,line2]).reset_index(drop=True)                


df_bk
```

```python
# Add 'Page From' column
import re
PageFrom = df_bk[bookname].copy()
for i in range(2,len(PageFrom)):
    if type(PageFrom[i]) != int:
        x = PageFrom[i]
        y = re.split(',|-',x)
        z = y[0]
        PageFrom[i]=z

df_bk['Page From'] = PageFrom
df_bk["Page From"] = pd.to_numeric(df_bk["Page From"])
df_bk
```

```python
# Add 'Page To' column
import re
PageTo = df_bk[bookname].copy()
for i in range(2,len(PageTo)):
    if PageTo[i] != 0:
        if type(PageTo[i]) == int:
            if int(PageFrom[i+1]) > PageTo[i]:
                    PageTo[i] = int(PageFrom[i+1])-1
        if type(PageTo[i]) != int:
            x = PageTo[i]
            y = x.split("-")
            z = y[-1]
            q = z.split(",")
            p = q[0]
            print(p)
            PageTo[i]=p

    else:
        PageTo[i] = 0

df_bk['Page To'] = PageTo
df_bk["Page To"] = pd.to_numeric(df_bk["Page To"])
df_bk
```

```python
# Add NonNumbered Pages
NonNumberedPages = df_bk[bookname].copy()
for i in range(len(NonNumberedPages)):
    if ',' in str(NonNumberedPages[i]):
        x = NonNumberedPages[i]
        y = x.split(",")
        z = y[-1]
        print(z)
        NonNumberedPages[i]=z

    else:
        NonNumberedPages[i] = 0

df_bk['NonNumberedPages'] = NonNumberedPages
df_bk["NonNumberedPages"] = pd.to_numeric(df_bk["NonNumberedPages"])
df_bk
```

```python
# Add No of Pages
NoofPages=[0,0]
for i in range(2,len(PageTo)):
    j = int(df_bk['Page To'][i]) - int(df_bk['Page From'][i]) + 1+int(df_bk['NonNumberedPages'][i])
    NoofPages.append(j)
    
df_bk['No of Pages'] = NoofPages
df_bk["No of Pages"] = pd.to_numeric(df_bk["No of Pages"])
df_bk
```

```python
# drop no of Front Cover & no of Back Cover
df_bk = df_bk.drop([0,1]).reset_index(drop=True)
df_bk
```
